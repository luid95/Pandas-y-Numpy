# If Condicion #

```python
# si intentamos dividir un numero un numero entre cero nos producira un errorZeroDivisionError
numero_1 = 100
numero_2 = 0 

division = numero_1/numero_2
# es infinito, por lo tanto produce errror
```

```python
# Si deseamos que nos avise cuando el denominador es 0. Podemos utilizar una condicion
numero_1 = 100
numero_2 = 0 

if(numero_2 != 0):
    division = numero_1 / numero_2
    print(division)

if(numero_2 == 0):
    print("Denominador = ", numero_2)
```

# If-else Condicion

```python
operacion='división'
valor1=13
valor2=2

if(operacion == 'suma'):
    print("La suma es : ", valor1+valor2)
elif(operacion == 'resta'):
    print("La resta es : ", valor1-valor2)
elif(operacion == 'multiplicacion' or operacion == 'multiplicación'):
    print("La multiplicacion es : ", valor1*valor2)
elif((operacion == 'division' or operacion == 'división') and valor2!=0):
    print("La division es : ", valor1/valor2)
else:
    print("Seleccione la operacion correcta; suma, resta, multiplicacion o division ")
```

# For bucle

```python
for num in range(1, 11): #mostrar los numeros del 1-1000
    print (num)
```

```python
for num in range(1, 1000):
    if(num%2 ==0):
        print ('\t', num) #mantener del lado derecho con tabulacion los numero pares
    else:
        print(num) #mostrar del lado izquierdo los numeros impares
```

```python
suma = 0
for num in range(1,1000):
    suma += num
print(suma)
```

```python
# Recorrer una lista
frutas =['manzana', 'banana', 'uva', 'pera']
for x in frutas:
    print (x)
```

```python
# Recorrer una cadena de texto
for x in 'El respeto al derecho ajeno es la paz':
    if(x!=' '): #imprimir sin espacios
        print(x)
```

```python
# Recorrer un diccionario
diccionario={'uno':1, 'dos':2, 'tres':3, 'cuatro':4}

for numeroletras, numero in diccionario.items():
    print('El signo ', numero,' se llama ', numeroletras)
```

# While bucle

```python
contar = 1
while(contar < 1000):
    print("La cuenta va por: ", contar)
    contar+=1
print("Termino...")
```

# Sentencia continue

La sentencia continue omite el resto del codigo dentro de un bucle solo para la iteraicon actual

```python
nuevaCadena = ''
for val in "cadena":
    if(val == 'd' or val=='e'):
        continue
    nuevaCadena+= val

    print(nuevaCadena)
```
